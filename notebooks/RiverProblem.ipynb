**5.1. River Problem** \
_An Exact Algorithm to make a Trade-off between Cost and Probability in SSPs_

> In the river problem an agent is in one side of the river bank and wants to go to the other side. The agent has two options: (i) swimming from any point of the river bank, or (ii) going along the river bank until a bridge. This problem is modeled as a grid Nx × Ny, where x = 1 and x = Nx represents the river bank; and y = Ny represents the bridge and y = 1 represents a waterfall where the agent can get trapped or die. The goal is in the other side of the river bank far from the bridge, xg = Nx and yg = 1. The agent can move in the cardinal directions. If actions are taken on the river bank or in the bridge then transitions are deterministic to the respective cardinal directions; if actions are taken in the river then transitions are probabilistic and follows the chosen cardinal directions with probability 1 − P robriver or follows down the river with probability P robriver. The waterfall is modeled as dead-end states. The immediate cost is 1. 

```python
def run_river_problem(river_flow):
    transition_probabilities = rp.build_transition_probabilities(grid_size, block_type, river_flow=river_flow)

    is_ok, dict_verification = rp._verify_sum_probabilities(transition_probabilities)
    if not is_ok: print(f'Soma das Probabilidades está válida? {is_ok}')

    rnp = RiskNeutralPlanning(grid_size, goal_state, transition_probabilities, costs, num_actions=4, discount_factor=discount_factor, epsilon=epsilon)

    num_iterations = rnp.run_converge()
    hm = viz_tools.visualize_V(rnp, rnp.V, grid_size, 4, goal_state, num_iterations, str_title=f'RiverProblem w/ ValueIteration - RiverFlow [{river_flow}]')
    
    return rnp
```

```python
viz_tools = VizTools()
rp = RiverProblem()

grid_size = (5, 20)
goal_state = (4, 0)
discount_factor = 0.95
epsilon = 0.001

costs = {0: -1, 1: -1, 2: -1, 3: -1}
block_type = rp.build_block_type(grid_size)
```

```python
rnp = run_river_problem(river_flow=0.5)
```
